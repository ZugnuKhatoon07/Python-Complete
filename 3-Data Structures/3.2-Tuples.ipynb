```python
# Tuple Methods
## count method check karta hai ki koi value tuple me kitna bar repeat hui hai
print(numbers.count(1))

## index method batata hai ki koi value tuple me sabse pahli baar kis position par aayi hai.
print(numbers.index(3))
```

```python
## Packing and Unpacking tuple
## packing
packed_tuple=1,"Hello",3.14
print(packed_tuple)
```

```python
## Unpacking a tuple
a,b,c=packed_tuple

print(a)
print(b)
print(c)
```

```python
## Unpacking with *
numbers = (1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

```python
nested_tuple = ((1,2,3), ("a","b","c"), (True, False))

## access the elements inside a tuple
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## iterating over nested tuples
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
```
