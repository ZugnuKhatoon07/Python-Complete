### Pandas-DataFrame And Series

Pandas is a powerful data manipulation library in Python. widely used for data analysis and data cleaning. it
provides two primary data structures: Seris and Dataframe. A Seris one-dimensional array-like object,
while a Dataframe is a two-dimensional, size-multable, and potentially heterogeneous tabular data structure
with labeled axes(rows and culumns.)

```python
## Series
## A pandas Series is a one-dimensional array-like object that can hold any data type. It is similar to a column in a table.

import pandas as pd 
data= [1,2,3,4,5]
series=pd.Series(data)
print("Series \n",series)
print(type(series))
```

```python
# Create a Series from dictionary
data={'a':1, 'b':2, 'c':3}
series_dict=pd.Series(data)
print(series_dict)
```

```python
data = [10,20,30]
index=['a', 'b', 'c']
pd.Series(data,index=index)
```

```python
# Dataframe
## create a Dataframe a dictionary of list

data ={
    'Name': ['Krish', 'John', 'Jack'],
    'Age': [25, 30, 45],
    'City': ['Banglore', 'New York', 'Florida']
}
df= pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```

```python
# Create a Data frame a List of Dictionaries

data=[
    {'Name':'Krish', 'Age':25, 'City':'Banglore'},
    {'Name':'John', 'Age':34, 'City':'Banglore'},
    {'Name':'Bappy', 'Age':25, 'City':'Banglore'},
    {'Name':'Jack', 'Age':25, 'City':'Banglore'}
]
df= pd.DataFrame(data)
print(df)
print(type(df))
```
