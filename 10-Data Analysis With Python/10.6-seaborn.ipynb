### Data Visualization With Seaborn

Seaborn is a Python Visualization library base on Matplotlib that provides a high-level interface for drawing
attractive and informative statiscal graphics. Seaborn help in creating complex visualization with just a few
lines of code. In this lesson, we will cover the basics of seaborn, including creating various types and plot
customizing them

```python
import seaborn as sns
```

```python
## Basic Plotting With Seaborn
tips=sns.load_dataset('tip')
tips
```

```python
## Create  a scatter with plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
```

```python
## Line Plot with Seaborn
sns.lineplot(x='size', y='total_bill', data=tips, color='green')
plt.title("Line Plot of Total Bill by Size of Party")
```

```python
## Category Plot with Seaborn
# Bar plot
sns.barplot(x='day', y='total_bill', data=tips, palette='pastel')
plt.title("Average Total Bill by Day")
plt.show()
```

```python
## Box Plot with Seaborn
sns.boxplot(x='day', y='total_bill', data=tips, palette='Set2')
```

```python
## violin plot

sns.violinplot(x='day', y='total_bill', data=tips, palette='muted')
```

```python
## Historgram with Seaborn
sns.histplot(tips['total_bill'], bins=10, kde=True, color='purple')
```

```python
## KDE plot 
sns.kdeplot(tips['total_bill'], fill=True, color='orange')
```

```python
## Pairplot with Seaborn
sns.pairplot(tips, hue='sex')
```

```python
## Heatmap with Seaborn
corr=tips[['total_bill', 'tip', 'size']].corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
import pandas as pd 
sales_df=pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
## plot sales data
plt.figure(figsize=(10,6))
sns.barplot(x='Product Category', y='Total Revenue', data=sales_df, estimator='sum')
plt.title("Total Revenue by Product Category")
plt.xticks(rotation=45)
plt.show()
```

```python
## plot total sales by Region 
plt.figure(figsize=(10,6))
sns.barplot(x='Region', y='Total Revenue', data=sales_df, estimator='sum')
plt.title("Total Revenue by Region")
plt.xticks(rotation=45)
plt.show()
```
