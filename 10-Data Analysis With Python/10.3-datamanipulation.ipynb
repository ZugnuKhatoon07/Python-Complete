```python
import pandas as pd
```

```python
## Handking Minssing V alues
df.isnull().any()
```

```python
df.isnull().sum()
```

```python
df_filled=df.fillna(0)
```

```python
## Filling missing values with the mean of the column
df['Total Revenue'] = df['Total Revenue'].fillna(df['Total Revenue'].mean())
df
```

```python
print(df.columns)
```

```python
df.dtypes
```

```python
# Renameing Columns
df=df.rename(columns={'Date': 'Sale Date'})
df.head()
```

```python
## changeing data types
df['Sale Date']= pd.to_datetime(df['Sale Date'])
df.head()
```

```python
df['Unit Price']=df['Unit Price'].fillna(df['Unit Price'].mean()).astype(int)
df.head()
```

```python
df['Unit Price']=df['Unit Price'].apply(lambda x:x*2)
df.head()
```

```python
# Data aggregating And Grouping
df_grouped=df.groupby('Product Category')['Total Revenue'].sum().reset_index()
df_grouped
```

```python
grouped_sum=df.groupby(['Product Category', 'Region'])['Total Revenue'].sum()
print(grouped_sum)
```

```python
grouped_sum=df.groupby(['Product Category', 'Region'])['Total Revenue'].mean()
print(grouped_sum)
```

```python
## aggreagate multiple functions
grouped_agg=df.groupby(['Product Category', 'Region'])['Total Revenue'].agg(['sum', 'mean', 'max', 'min'])
print(grouped_agg)
```

```python
## Merging and Joining DataFrames
df1=pd.DataFrame({
    'Product ID': [1, 2, 3, 4, 5],
    'Product Name': ['A', 'B', 'C', 'D', 'E']
})

df2=pd.DataFrame({
    'Product ID': [1, 2, 3, 5], 
    'Price': [100, 200, 300, 400]
})
```

```python
df1,df2
```

```python
## Merging on the 'Key Column' 
merged_df=pd.merge(df1, df2, on='Product ID', how='inner')
merged_df
```

```python
pd.merge(df1, df2, on='Product ID', how='outer')
```

```python
pd.merge(df1, df2, on='Product ID', how='left')
```

```python
pd.merge(df1, df2, on='Product ID', how='right')
```
