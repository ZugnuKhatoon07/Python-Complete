### Assignment 6: Working with the `os` Module

Use the `os` module to create a new directory, list the contents of the current directory, and remove the newly created directory.


```python
import os

folder = "new_directory"

# Create directory only if it doesn't exist
if not os.path.exists(folder):
    os.mkdir(folder)
    print("üìÅ Directory created.")
else:
    print("‚ö†Ô∏è Directory already exists.")

print("Before removing:", os.listdir('.'))

# Remove directory only if it exists
if os.path.exists(folder):
    os.rmdir(folder)
    print("üóëÔ∏è Directory removed.")
else:
    print("‚ö†Ô∏è Directory not found.")

print("After removing:", os.listdir('.'))

```
