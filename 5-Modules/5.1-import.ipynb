##### Importing in Python: Modules and Packages

#### In Python, modules and packages help organiz and reuse code. Here's a comrehensive guide on how to import them

```python
import math
math.sqrt(16)
```

```python
from math import sqrt,pi
print(sqrt(16))
print(sqrt(25))
print(pi)
```

```python
import numpy as np
np.array([1, 2, 3, 4])
```

```python
from math import *
print(sqrt(16))
print(pi)
```

```python
from package.maths import addition
addition(2,3)
```

```python
from package import maths
maths.addition(2,3)
```
