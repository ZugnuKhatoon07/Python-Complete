```python
## creating a function
def even_or_odd(num):
    """ Check if a number is even or odd."""
    if num % 2 == 0:
        print(f"The number {num} is even.")
    else:
       print(f"The number {num} is odd.")
```

```python
## Call thisfuction
even_or_odd(24)
```

```python
## Default Parameters

def greet(name):
    print(f"Hello {name} Welcome to the paradise!")

greet("Zugnu")
```

```python
def greet(name="Guest"):
    print(f"Hello {name} Welcome to the paradise!")

greet("Sruu")
```

```python
### Variable Length Aeguments
## Position and keywords  arguments

def print_numbers(*Zugnu):
    """ Print numbers passed as arguments."""
    for number in Zugnu:
        print(number)
```

```python
print_numbers(1, 2, 3, 4, 5, 6,"Zugnu")
```

```python
# Positional arguments 
def print_numbers(*args):
    """ Print numbers passed as arguments."""
    for number in args:
        print(number)
```

```python
print_numbers(1, 3, 4, 5, 6, "Zugnu")
```

```python
### keyword Aeguments

def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f" {key}: {value}")
```

```python
print_details(name="zugnu", age=24, city="Siwan", country="bihar")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f"Positional arument :{val}")

    for key, value in kwargs.items():
        print(f" {key}: {value}")
```

```python
print_details(1, 2, 3, 4, 5, name="zugnu", age=24, city="Siwan", country="bihar" )
```
