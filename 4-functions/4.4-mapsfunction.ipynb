```python
def square(x):
    return x ** x
square(2)
```

```python
numbers = [1, 2, 3, 4, 5]
list(map(square, numbers))
```

```python
## Lambda function with map
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
list(map(lambda x:x*x, numbers))
```

```python
## map() multiple iterables
numbers1 = [1, 2, 3, 4]
numbers2 = [5, 6, 7, 8]
added_numbers = list(map(lambda x, y: x + y, numbers1, numbers2))
print(added_numbers)
```

```python
## map() to covert a list of strings to integers
## use map to covert strings to integers
str_numbers = ['1', '2', '3', '4', '5']
int_numbers = list(map(int, str_numbers))
print(int_numbers)
```

```python
words = ['apple', 'banana', 'cherry']
upper_words = list(map(str.upper, words))
print(upper_words)
```

```python
def get_name(person):
    return person['name']
people = [
    {'name': 'Zugnu', 'age': 20},
    {'name': 'Sarfaraz', 'age': 20}
]

list(map(get_name, people))
```
