```python
# Syntax
lambda arguments : expression
```

```python
def addition(a,b):
    return a+b

addition(2,3)


    
```

```python
addition = lambda a,b: a+b 
type(addition)
print(addition (3,4))
```

```python
def even(num):
    if num % 2 == 0:
        return True
even(24)
    
    
```

```python
even1=lambda num:num%2==0
even1(12)
```

```python
def addition(x, y, z):
    return x + y + z

addition(1, 2, 3)
```

```python
addition = lambda x, y, z: x+y+z
addition(2,3,4)
```

```python
##map() - applies a function to all items in an iterable (list, tuple, ets)
numbers = [1, 2, 3, 4, 5]
def square(number):
    return number ** 2
square(2)
```

```python
list(map(lambda x:x**2, numbers))
```
