Saare values ek saath memory me store hote hain.

```python
def normal_func():
    return [1, 2, 3, 4, 5]

print(normal_func())

##Yahan sabhi numbers ek saath list me memory me store ho gaye.
```

```python
def square(n):
    for i in range(3):
        yield i ** 2
```

## üëâ Generator Function:

yield keyword use karta hai.

Har baar yield ek value deta hai aur function temporarily ruk jata hai.

Jab next value chahiye hoti hai, function wahi se dobara start hota hai.

Memory me sab values ek saath store nahi hoti ‚Äî ek-ek karke milti hain.

```python
def gen_func():
    for i in range(1, 6):
        yield i  # return nahi, yield use kiya hai

g = gen_func()
print(next(g))  # Output: 1
print(next(g))  # Output: 2
```

```python
def square(n):
    for i in range(3):
        yield i**2
```

```python
square(3)
```

```python
for i in square(3):
    print(i)
```

```python
a = square(3)
a
```

```python
next(a)
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
```

```python
gen=my_generator()
gen
```

```python
next(gen)
```

```python
for val in gen:
     print(val)
```

### Practical Example: Reading Large Files 
Generators are particularly useful for reading large files because they allow you to proccess one 
line at a time without loading the entire file into memory. 

```python
### Pacrtical : Reading Large files

def read_large_file (file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line


```

```python
file_path='large_file.txt'

for line in read_large_file(file_path):
    print(line.strip())
```
