### Decorators

Decorators are a powerful and flexible feature in python that allows you to modify the behavior of a function or class 
method. They are commonly used to add functionality to functions or methods without modifying theai actual code.
this lesson covers the basics of decorators , including how to create and use them.

```python
## Decorator
def main_welcome(func):
    
    def sub_welcome_method():
        print("Welcome to the advance python course")
        func( )
        print("Please learn these concepts properly")
    return sub_welcome_method()
```

```python
def coure_introduction():
    print("This is an advanced python course")

coure_introduction()
```

```python
main_welcome(coure_introduction)
```

```python
@main_welcome
def coure_introduction():
    print("This is an advanced python course")
```

```python
## Decorator

def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper
```

```python
@my_decorator
def say_hello():
    print("Hello!")
```

```python
say_hello()
```

```python
## Decorator with arguments
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print("Hello!")
```

```python
say_hello()
```
