# Iterators
iterators are advanced Python concepts that allow for effocient looping and memroy management. iterators provide a way to access elements of a collection sequentially without exposing the underlying structure.

```python
my_list=[1,2,3,4,5,6]
for i in my_list:
    print(i)
```

```python
type(my_list)
```

```python
print(my_list)
```

```python
## Iterator
iterator=iter(my_list)
print(type(iterator))
```

```python
iterator
```

```python
## iterate through all the element

next(iterator)
```

```python
iterator=iter(my_list)
```

```python

try:
    print(next(iterator))
except StopIteration:
    print("There are no elements in the iterator")
```

```python
# Strig iterator
my_string = "Hello"
string_iterator = iter(my_string)


print(next(string_iterator)) # Output: H
print(next(string_iterator)) # Output: e
print(next(string_iterator)) # Output: o
print(next(string_iterator)) # Output: l
print(next(string_iterator)) # Output: l
print(next(string_iterator)) # Output: l
```
