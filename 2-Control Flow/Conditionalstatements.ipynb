### Conditional Statements(if, elif, else)

```python
## if statment
age = 20
if age >= 18:
    print("You are allowed to vote in the elections")
```

```python
age>=18
```

```python
## else 
## The else statement executes a block of code of the condition in the if statement is False.
age = 16

if age >= 18:
    print("You are eligiable for voting")
else:
    print("You are a minor")
```

```python
## elif
## The elif ststement allows you to check multiple conditions. It stands for "else if"

age = 20

if age<13:
    print("You are a child")
elif age<18:
    print("You are a teenager")
else:
    print("You are an adult")
```

```python
## Nested Conditional Statements

# You can place one or monr if, elif, or else satatments inside another if, elif, or else statement to create nested conditional statements.

## number even, odd, negative

num= int(input("Enter the number"))

if num>0:
    print("This number is positive")
    if num%2 == 0:
        print("The number is even")
    else:
        print("This number is odd")
else:
    print("This number is zero or negative")
```

```python
## Practical Examples

## Determine if a year is a leap year using nested condition statement

year=int(input("Enter the year"))

if (year%4==0):
    if year % 100 == 0:
        if year % 400 == 0:
            print(year,"is a leap year")
        else:
            print(year,"is not a leap year")
    else:
        print(year,"is a leap year")
else:
    print(year,"is not a leap year")
```

```python
## Assignment
## Simple Calculator prongram
# Take user input
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operation = input("Enter Operation (+, -, *, /): ")

# Perform calculation based on user input
if operation == '+':
    result = num1 + num2
elif operation == '-':
    result = num1 - num2
elif operation == '*':
    result = num1 * num2
elif operation == '/':
    if num2 != 0:
        result = num1 / num2
    else:
        result = "Error! Division by zero."
else:
    result = "Invalid operation: "
print("Result:", result)
```
