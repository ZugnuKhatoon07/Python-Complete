# Understanding Exception

#### Exception handling in python allows you to handle errors gracefully and take corrective actions without stopping the exection of the program. this lesson will cover the basic of exception, including how to use try,except, else, and finally blocks.

# What Are Exception?

#### Exceptions are events that disrupt the normal flow of a program. They occur when an error is encountered during program execution. Common exceptions include:

#### ZeroDivisionError:Dividing by Zero.
#### FileNotFoundError:File not found.
#### ValueError:Invalid value.
#### TypeError:Invalid type

```python
## Exception try , exception block

try:
    a=b 
except:
    print("The variable has not been assigned")
```

```python
try:
    num=int(input("Enter a number"))
    result=10/num
except ValueError:
    print("This in not a valid number")
except ZeroDivisionError:
    print("enter denominator greater than 0")
```

```python

## try,except,else block
try:
    num=int(input("Enter a number:"))
    result=10/num
except ValueError:
    print("That's not a valid number!")
except ZeroDivisionError:
    print("You can't divide by zero!")
except Exception as ex:
    print(ex)
else:
    print(f"the result is {result}")
```

```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ValueError:
    print("That's not a valid number!")
except ZeroDivisionError:
    print("You can't divide by zero!")
except Exception as ex:
    print(ex)
else:
    print(f"The result is {result}")
finally:
    print("Execution complete.")
```
